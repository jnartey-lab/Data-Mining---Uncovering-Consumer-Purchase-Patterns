## 3.Feature Engineering
    - Create new features 

```python
#  Transaction Date Features
online_retail_df['InvoiceDate'] = pd.to_datetime(online_retail_df['InvoiceDate'], dayfirst=True)

# Total Revenue per Transaction
online_retail_df['TotalPrice'] = online_retail_df['Quantity'] * online_retail_df['UnitPrice']

# Number of Items per Invoice
invoice_item_counts = online_retail_df.groupby('InvoiceNo')['Quantity'].sum().reset_index(name='TotalItems')
online_retail_df = online_retail_df.merge(invoice_item_counts, on='InvoiceNo')
#display(online_retail_df)


# Customer Spending Behavior
customer_spending = online_retail_df.groupby('CustomerID')['UnitPrice'].sum().reset_index(name='TotalSpent')
online_retail_df = online_retail_df.merge(customer_spending, on='CustomerID')
#display(online_retail_df)
```

```python
# For Cohort Analysis
online_retail_df['InvoiceDate'] = pd.to_datetime(online_retail_df['InvoiceDate'], errors='coerce')

online_retail_df['CohortMonth'] = online_retail_df['InvoiceDate'].dt.to_period('M')
first_purchase = online_retail_df.groupby('CustomerID')['InvoiceDate'].min().dt.to_period('M')
online_retail_df['CohortGroup'] = online_retail_df['CustomerID'].map(first_purchase)

monthly_sales = online_retail_df.set_index('InvoiceDate').resample('M')['UnitPrice'].sum()

import matplotlib.pyplot as plt

monthly_sales.plot(figsize=(12,6), title='Monthly Sales Trend')
plt.ylabel('Unit Sales')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```

```python
# --- RFM Feature Engineering ---
snapshot_date = online_retail_df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = online_retail_df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                   # Frequency
    'TotalPrice': 'sum'                                       # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
rfm = rfm[rfm['Monetary'] > 0]  # Remove customers with 0 total spend just in case

# --- Scaling ---
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])
```

## Part 4. Machine learning
        - Apriori Algorithm 
        - FP-Growth Algorithm
       

```python
# Filter out items that appear in fewer than 50 transactions and sample a random subset

item_counts = online_retail_df['Description'].value_counts()
common_items = item_counts[item_counts > 50].index 

filtered_df = online_retail_df[online_retail_df['Description'].isin(common_items)]

sampled_invoices = filtered_df['InvoiceNo'].drop_duplicates().sample(n=3000, random_state=42)
sampled_df = filtered_df[filtered_df['InvoiceNo'].isin(sampled_invoices)]

```

## Applying Apriori Algorithm 

```python
basket = sampled_df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0)

# Convert to binary (presence/absence)
basket_sets = basket.applymap(lambda x: 1 if x > 0 else 0)

from mlxtend.frequent_patterns import apriori, association_rules

# Get frequent itemsets with min_support
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules = rules.sort_values(by='lift', ascending=False)

rules.head()
```

```python
# Display top 10 rules
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```

```python
# Filter for single-item itemsets
single_items = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) == 1)]

# Sort and take top 10
top_items = single_items.sort_values(by='support', ascending=False).head(10)

# Convert frozensets to string for plotting
top_items['item'] = top_items['itemsets'].apply(lambda x: list(x)[0])

# Plot
plt.figure(figsize=(10,6))
plt.barh(top_items['item'][::-1], top_items['support'][::-1])
plt.xlabel('Support')
plt.title('Top 10 Most Frequent Items')
plt.show()
```

```python
# Filter for itemsets with 2 or more items
multi_items = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) >= 2)]

# Take top 10
top_itemsets = multi_items.sort_values(by='support', ascending=False).head(10)

# Convert frozensets to strings
top_itemsets['itemset_str'] = top_itemsets['itemsets'].apply(lambda x: ', '.join(x))

# Plot
plt.figure(figsize=(10,6))
plt.barh(top_itemsets['itemset_str'][::-1], top_itemsets['support'][::-1])
plt.xlabel('Support')
plt.title('Top 10 Frequent Itemsets (2+ items)')
plt.show()
```

```python
import seaborn as sns
import numpy as np

# Create basket again for co-occurrence
basket_binary = basket_sets.loc[:, basket_sets.sum().sort_values(ascending=False).head(15).index]

# Compute co-occurrence
co_matrix = np.dot(basket_binary.T, basket_binary)

# Convert to DataFrame
co_df = pd.DataFrame(co_matrix, index=basket_binary.columns, columns=basket_binary.columns)

# Plot heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(co_df, cmap="Blues", annot=True, fmt='d')
plt.title('Co-occurrence of Top 15 Items')
plt.show()
```

# Applying FP Growth

```python
from mlxtend.frequent_patterns import fpgrowth

# Run FP-Growth
frequent_itemsets_fp = fpgrowth(basket_sets, min_support=0.02, use_colnames=True)

# Sort by support for inspection
frequent_itemsets_fp = frequent_itemsets_fp.sort_values(by='support', ascending=False)

# View top itemsets
frequent_itemsets_fp.head(10)
```

```python
# Filter for single-item itemsets
top_fp_items = frequent_itemsets_fp[frequent_itemsets_fp['itemsets'].apply(lambda x: len(x) == 1)].head(10)

# Convert frozensets to string for plotting
top_fp_items['item'] = top_fp_items['itemsets'].apply(lambda x: list(x)[0])

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(top_fp_items['item'][::-1], top_fp_items['support'][::-1])
plt.xlabel('Support')
plt.title('Top 10 Items from FP-Growth')
plt.show()
```

```python
from mlxtend.frequent_patterns import association_rules

rules_fp = association_rules(frequent_itemsets_fp, metric="lift", min_threshold=1)
rules_fp = rules_fp.sort_values(by='lift', ascending=False)

# View top rules
rules_fp[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```

# Calculate more metrics to inspect the association rules

```python
# Calculate additional metrics
rules['leverage'] = rules['support'] - (rules['antecedent support'] * rules['consequent support'])

rules['conviction'] = (1 - rules['consequent support']) / (1 - rules['confidence'])

rules['kulczynski'] = 0.5 * (rules['confidence'] + rules['support'] / rules['consequent support'])

rules['jaccard'] = rules['support'] / (rules['antecedent support'] + rules['consequent support'] - rules['support'])
```

```python
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift', 
       'leverage', 'conviction', 'kulczynski', 'jaccard']].sort_values(
    by='kulczynski', ascending=False).head(10)
```

# Findings:

**Consistent Patterns Across Algorithms:**

By using both the Apriori and FP-Growth Algorithms, strong purchasing patterns in the retail dataset were identified. The consistency of these patterns across both methods demonstrates the robustness of the itemsets and association rules found.

**Analysis of Rules:**

Various metrics, such as support, confidence, lift, leverage, conviction, Kulczynski, and Jaccard, were used to evaluate the association rules. These metrics helped to clarify the strength and significance of the identified relationships between products.

**Key Product Findings:**

Teacup products emerged as the strongest rules, with high scores across all metrics. The rules associated with teacups indicated that when customers buy a specific color of teacup, they are very likely to purchase another color variation from the same series.

Lift values over 20, confidence above 80%, Kulczynski scores around 0.73, and conviction values reaching up to 10.29 confirm the strength of these associations.

Similar patterns were observed for products with multiple variations, such as lunchboxes and alarm clocks, which also have different color options. Seasonal items like Christmas décor products also showed strong purchasing patterns.

**Product Grouping:**

To encourage pattern behavior, group similar products physically in stores or digitally on e-commerce websites. Since the discovered rules are bi-directional, displaying items like teacups together increases the likelihood of them being bought together.

**Stock Management:**

Ensure that sufficient stock levels are maintained for popular products, especially during peak seasons or promotional periods, to meet the demand generated by these strong purchasing patterns.

Having adequate stock can help prevent potential losses due to product unavailability, especially for high-demand items like teacup sets, lunchboxes, and seasonal products.
